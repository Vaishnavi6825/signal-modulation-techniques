GENERATION OF ANALOG SIGNAL <br>
PROGRAM:

```
import numpy as np
import matplotlib.pyplot as plt

# Generate a low-frequency analog signal (sine wave) over a short duration
time = np.linspace(0, 0.1, 100) # Shorter time axis (0 to 0.1 seconds)
frequency = 15  # Frequency of 1 Hz
analog_signal = np.sin(2 * np.pi * frequency * time)

# Plotting the low-frequency analog signal
plt.figure(figsize=(10,4))
plt.plot(time, analog_signal, color='blue', label='Analog Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Analog Signal')
plt.legend()
plt.show()
```


DELTA MODULATION(DM): <br>

PROGRAM:


```
import numpy as np
import matplotlib.pyplot as plt

# Generate a low-frequency analog signal (sine wave) over a short duration
time = np.linspace(0, 0.1, 100)
frequency = 15  # Frequency of 15 Hz
analog_signal = np.sin(2 * np.pi * frequency * time)

# Delta Modulation Parameters
step_size = 0.1  # Step size for DM
modulated_signal = np.zeros_like(analog_signal)

# Initial condition
modulated_signal[0] = analog_signal[0]

# Delta modulation loop
for i in range(1, len(time)):
    if analog_signal[i] > modulated_signal[i-1]:
        modulated_signal[i] = modulated_signal[i-1] + step_size
    else:
        modulated_signal[i] = modulated_signal[i-1] - step_size

# Plotting the Delta Modulated signal plt.figure(figsize=(10,4))
plt.step(time, modulated_signal, where='mid', color='green', label='Delta Modulated Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Delta Modulation (DM)')
plt.legend()
plt.show()
```

ADAPTIVE DELTA MODULATION(ADM): <br>

PROGRAM:


```
import numpy as np
import matplotlib.pyplot as plt

# Generate a low-frequency analog signal (sine wave) over a short duration
time = np.linspace(0, 0.1, 100)
frequency = 15  # Frequency of 15 Hz
analog_signal = np.sin(2 * np.pi * frequency * time)

# Adaptive Delta Modulation Parameters
modulated_signal = np.zeros_like(analog_signal)
step_size = 0.1  # Initial step size
adaptive_step_size = np.zeros_like(analog_signal)

# Initial condition
modulated_signal[0] = analog_signal[0]

# Adaptive delta modulation loop
for i in range(1, len(time)):
    diff = analog_signal[i] - modulated_signal[i-1]

    # Adjust step size based on the difference
    adaptive_step_size[i] = step_size * (1 + abs(diff))  # Increase step size if difference is large

    # Modulation logic
    if diff > 0:
        modulated_signal[i] = modulated_signal[i-1] + adaptive_step_size[i]
    else:
        modulated_signal[i] = modulated_signal[i-1] - adaptive_step_size[i]

# Plotting the Adaptive Delta Modulated signal
plt.figure(figsize=(10, 4))
plt.step(time, modulated_signal, where='mid', color='blue', label='Adaptive Delta Modulated Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Adaptive Delta Modulation (ADM)')
plt.legend()
plt.grid(True)  # Optional: Adds a grid for better visualization
plt.show()


```

PULSE CODE MODULATION(PCM): <br>
PROGRAM:


```
import numpy as np
import matplotlib.pyplot as plt

# Generate a low-frequency analog signal (sine wave) over a short duration
time = np.linspace(0, 0.1, 100)
frequency = 15  # Frequency of 15 Hz
analog_signal = np.sin(2 * np.pi * frequency * time)

# PCM: Sample and quantize the analog signal
num_samples= 8  # Number of quantization levels
quantized_signal = np.round((analog_signal - np.min(analog_signal)) / (np.max(analog_signal) - np.min(analog_signal)) * (num_samples - 1)) # Quantized signal

# Plotting the PCM signal (Quantized Digital Signal)
plt.figure(figsize=(10, 4))
plt.step(time, quantized_signal, where='mid', color='purple', label='PCM Quantized Signal')
plt.xlabel('Time')
plt.ylabel('Quantized Amplitude')
plt.title('Pulse Code Modulation (PCM)')
plt.legend()
plt.show()
```
